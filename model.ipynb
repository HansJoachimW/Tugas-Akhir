```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO
```

```python
api = KaggleApi()
api.authenticate()
```

```python
dataset_path = "dataset"
dataset_name = "pythonafroz/solar-panel-images"
```

```python
if not os.path.exists(dataset_path):
    print("Dataset not found. Downloading...")
    api.dataset_download_files(dataset_name, path=dataset_path, unzip=True)
    print("Download completed. Dataset files are stored at:", dataset_path)
else:
    print("Dataset already exists at:", dataset_path)
```

```python
for root, dirs, files in os.walk(dataset_path):
    print(f"Root: {root}")
    # print(f"Dirs: {dirs}")
    # print(f"Files: {files}")
```

```python
image_path = os.path.join(dataset_path, 'Faulty_solar_panel\Dusty\Dust (41).jpg')
```

```python
def showImage(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
showImage(image)
```

```python
def preprocess_image(image_path, target_size=(416, 416)):
    # Read image
    image = cv2.imread(image_path)
    # Resize image to target size
    image_resized = cv2.resize(image, target_size)
    # Normalize image
    image_normalized = image_resized / 255.0
    return image_normalized
```

```python
image_preprocessed = preprocess_image(image_path)
showImage(image_preprocessed)
```
